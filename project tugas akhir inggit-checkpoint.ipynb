Fungsi untuk mengisi data yang hilang dengan rata-rata

```python
def isi_data_hilang_dengan_rata_rata(df):
    for kolom in df.columns:
        if pd.api.types.is_numeric_dtype(df[kolom].dtype):  # Cek tipe data kolom
            rata_rata = df[kolom].mean()
            df[kolom].fillna(rata_rata, inplace=True)
```

Memanggil fungsi untuk mengisi data yang hilang dengan rata-rata

```python
isi_data_hilang_dengan_rata_rata(df)
```

# Metode Elbow untuk Tingkat Kemiskinan di Pulau Jawa

 mengambil nilai dari kolom 'Jumlah penduduk miskin' (000)

```python
X = df[['Presentase penduduk miskin']].values
```

membuat sebuah list (daftar) dengan nama wcss

```python
wcss = []
```

 menghitung nilai Within-Cluster Sum of Squares (WCSS) untuk berbagai jumlah cluster (dari 1 hingga 10).

```python
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

membuat sebuah grafik yang menampilkan hasil dari metode elbow (siku) dalam menentukan jumlah cluster yang optimal dalam menentukan jumlah cluster

```python
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='b')
plt.title('Metode Elbow')
plt.xlabel('Jumlah Cluster')
plt.ylabel('WCSS (Within-cluster sum of squares)')
plt.grid(True)
plt.show()
```
